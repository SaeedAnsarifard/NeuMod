```python
import sys
import os

import numpy as np
from numpy import arctan, arcsin
from math import sqrt, pi


# Get the current working directory
peanuts_path = os.path.abspath(os.path.join(os.getcwd(), "external", "PEANUTS"))

# Add PEANUTS to the Python path
sys.path.append(peanuts_path)


from peanuts.pmns import PMNS
from peanuts.solar import SolarModel, solar_flux_mass, Psolar
from peanuts.earth import EarthDensity, Pearth
from peanuts.evolutor import FullEvolutor


from math import radians
from peanuts.time_average import NadirExposure


import matplotlib.pyplot as plt

plt.rcParams['text.usetex']=True
plt.style.use('classic')
```

```python
ECCENTRICITY = 0.016698
ASTRO_UNIT    =  1.496 #1e11 m
RHO_DM2  = np.sqrt(2 * 0.4 * 7.65) #e-21 GeV^2
LIGHT_SPEED = 2.998 # 1e8 m/s


def SunEarthDistance(theta1=0, theta2= 2 * pi, resolution=0.002):
    theta = np.arange(theta1,theta2,resolution)
    cos   = np.cos(theta)
    l     = (1 - ECCENTRICITY**2) / (1 + ECCENTRICITY * cos)   
    return l,theta

param ={'T12'  : 0.308 ,
        'T13'  : 8.57, 
        'mu1'  : 0.,
        'mu2'  : 0., 
        'mu3'  : 0.  ,
        'mdm'  : 1e3 , 
        'alpha': 0.  ,
        'epsx' : 0.  ,
        'epsy' : 0.  ,
        'M12'  : 7.5e-5 }

ls,th_radians = SunEarthDistance()
ls_meters = ls * ASTRO_UNIT

norm_time = np.trapz(ls_meters**2, th_radians)
delta_time = []
cond = np.full(ls.shape, False)
for i in range (len(ls)):
    delta_t = np.trapz(ls_meters[:i]**2, th_radians[:i]) 
    if delta_t != 0:
        delta_time.append(delta_t / norm_time)
        cond[i] = True

delta_time = np.array(delta_time)

ls_meters  = ls_meters[cond]
th_radians = th_radians[cond]
year_list  = delta_time * 3.6525 * 2.4 * 6. * 6. / 6.6 # in 1e21 eV^-1
```

```python
def ULDMTerms(param):
    polar_vec = np.sqrt((1 - ( param['epsx'] * np.cos(th_radians) + param['epsy'] * np.sin(th_radians))**2))
    
    mass_var = 1e9 * (RHO_DM2/param['mdm']) * (np.cos(param['mdm'] * year_list + np.radians(param['alpha'])) 
                                      - np.cos(param['mdm'] * (year_list + (1e-2/6.6) * ls_meters/LIGHT_SPEED) + np.radians(param['alpha'])))
    
    
    uldm_term1 = np.cos(param['mu1'] * polar_vec * mass_var)**2 / (ls_meters/ASTRO_UNIT)**2 
    uldm_term2 = np.cos(param['mu2'] * polar_vec * mass_var)**2 / (ls_meters/ASTRO_UNIT)**2 
    uldm_term3 = np.cos(param['mu3'] * polar_vec * mass_var)**2 / (ls_meters/ASTRO_UNIT)**2 
    return np.array([uldm_term1,uldm_term2,uldm_term3])
```

```python
#1e-25 to 1e-17
param['mu1'] = 0.
param['mu2'] = 0.
param['mdm'] = 1e4
param['epsx'] = 0.
param['epsy'] = 1.
uldm_term = ULDMTerms(param)

print(uldm_term.shape)
plt.plot(th_radians, uldm_term[0],'b')
plt.plot(th_radians, uldm_term[1],'r')
plt.plot(th_radians, uldm_term[2],'g')
```

```python
day_time = np.int16(delta_time * 365.25)
hour_time = np.mod(delta_time * 365.25, day_time) * 24.
hour_time[day_time<1] = delta_time[day_time<1] * 365.25 * 24.

sin_deltas = -np.sin(0.491) * np.cos(day_time * 2 * pi / 365)
detector_latitude = 36
cos_eta  = np.cos(np.radians(detector_latitude)) * np.cos(hour_time * 2 * pi / 24) * np.sqrt(1-sin_deltas**2) - np.sin(np.radians(detector_latitude)) * sin_deltas
eta_list = np.arccos(cos_eta)
```
